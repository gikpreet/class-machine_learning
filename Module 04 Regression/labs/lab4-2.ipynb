# Lab 4-2 회귀 직선 구하기

## 라이브러리 import

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 데이터 준비

```python
hours = [2,3,4,5,6,7,8]
score = [24,40,35,60,75,75,90]
```

```python
X = np.array(hours)
y = np.array(score)
```

## 데이터 시각화

```python
#plt.scatter(x=y, y=X)
plt.scatter(X, y)
plt.xlabel('hours (독립 변수)')
plt.ylabel('score (종속 변수)')
plt.show()
```

## 기울기와 절편 구하기

### 기울기 구하기

```python
N = len(X)  # 데이터 포인트의 개수
    
# 기울기 계산 공식 적용
numerator = N * np.sum(X * y) - np.sum(X) * np.sum(y)
denominator = N * np.sum(X**2) - (np.sum(X))**2
    
m = numerator / denominator
print(m)
```

### 절편 구하기

```python
N = len(X)
b = (np.sum(y) - m * np.sum(X)) / N

print(b)
```

## 회귀 직선에 따른 예측 값 계산
각 독립 변수에 따른 종속 변수의 값 계산

```python
y_pred = m * X + b
print(y_pred)
```

## 회귀 곡선

```python
plt.scatter(X, y)
plt.plot(X, y_pred, color='red',)
plt.xlabel('hours (독립 변수)')
plt.ylabel('score (종속 변수)')
plt.show()
```

## 예측

```python
new_hour = 10
y_pred = b + m * new_hour

print(y_pred)
```
